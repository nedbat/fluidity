```python
svg_row(
    *[
        Fluidity(LinearNoise(seed=s), npoints=10, nlines=1, curver=hobby_curve, sorter=HilbertSortFirstLine())
            .draw(size=(200,200), curve_width=0.5, point_color=(1, 0, 0, 1), point_size=2, line_color=(0, 1, 0, 1), line_width=1)
        for s in range(5)
    ]
)
```

```python
N = 20
COLS = 4
ROWS = N // COLS
fs = [
    Fluidity(LinearNoise(seed=s), npoints=10, nlines=1, curver=hobby_curve, sorter=HilbertSortFirstLine())
    for s in range(N)
]
with cairo_context(200 * COLS, 200 * ROWS) as ctx:
    for f, tctx in zip(fs, context_tiles(ctx, rows=ROWS, cols=COLS)):
        f.draw_in_context(tctx, curve_width=0.5, point_color=(1, 0, 0, 1), point_size=2, line_color=(0, 1, 0, 1), line_width=1)
ctx
```
