```python
from fluidity import Fluidity, LinearNoise
f = Fluidity(LinearNoise(seed=55, istep=0.003), npoints=15, nlines=50)
f.draw(size=(400, 400))
```

```python
f1 = f.tweak(one_order=True)
f1.draw(size=(400, 400))
```

```python
Fluidity(LinearNoise(seed=55, istep=0.003), npoints=15, nlines=50, one_order=True).draw(size=(400, 400))
```

```python
Fluidity(LinearNoise(seed=55, istep=0.003, istart=0.07), npoints=15, nlines=50, one_order=True).draw(size=(400, 400))
```
